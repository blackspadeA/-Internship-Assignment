```python
name = input("Enter customer name: ")
units = int(input("Enter electricity units consumed: "))

if units <= 100:
    bill = units * 5
else:
    bill = (100 * 5) + ((units - 100) * 8)

print(f"Customer Name : {name}")
print(f"Units Consumed : {units}")
print(f"Total : ₹{bill}")
```
